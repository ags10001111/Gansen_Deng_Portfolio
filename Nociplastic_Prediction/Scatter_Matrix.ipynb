```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
file_path = "df_pre_orig.csv"
df = pd.read_csv(file_path)

# Drop unnecessary index column if it exists
df = df.drop(columns=["Unnamed: 0"], errors='ignore')

# Define color mapping for Nociplastic variable (0: Black, 1: Red)
palette = {0: "black", 1: "red"}

# Create improved scatterplot matrix
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue="Nociplastic", diag_kind="hist", palette=palette, markers=["o", "s"], diag_kws={'bins': 10, 'multiple': 'stack'})

# Add a title
#plt.suptitle("Scatterplot Matrix of APD, AND, MPD, and CSI\n(Color represents Nociplastic variable)", 
            # y=1.02, fontsize=14)

# Show the plot
plt.show()
```
